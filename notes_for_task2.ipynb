# BankFive Preprocessing Notes

## Noise Removal
I specifically targeted the `…` pattern because Trustpilot repeats the "Review Title" inside the "Review Body". Keeping both would double the weight of words in the title, biasing the sentiment analysis.

## Subject Tagging (Regex)
In banking, the "Subject" matters immensely. Sentiment about a "Loan" (high-stakes, long-term) is analyzed differently than sentiment about an "ATM" (transactional). This partitioning allows us to see if the bank is failing in digital tools but winning in face-to-face service.

## SymSpell vs TextBlob for Spelling
I chose **SymSpell** because it is significantly faster for large datasets. It uses a "Symmetric Delete" algorithm that can handle 140+ records in milliseconds.

## Lemmatization
I used **Lemmatization** instead of Stemming. Stemming would chop "Banking" to "Bank", but it might chop "Friendly" to "Friendl". Lemmatization ensures "Friendly" remains a meaningful root for sentiment.

## No Stopword Removal (By Choice)
I did not include a mandatory stopword removal flag because, in sentiment analysis, words like "not" or "never" (which are often in stopword lists) are vital. Removing them turns "not good" into "good", which destroys the sentiment accuracy.
